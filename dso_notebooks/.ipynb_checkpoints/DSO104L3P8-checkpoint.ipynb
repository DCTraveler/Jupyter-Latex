Sometimes we will need to convert a categorical variable into a matrix. Many times for statistical modeling and or machine learning model development we need to create dummy variables.

Pandas has a function which can turn a categorical variable into a series of zeros and ones, which makes them a lot easier to quantify and compare.

pandas has a function for dummy coding in Python, called pd.get_dummies. You can feed this function the dataset name and variable name for the column you want to dummy code, and it will create a data frame with the columns for that variable all set up as zeroes and ones.

Let's say we want to create a matrix on which way the superhero leans (good, bad, or neutral).

We'll create a matrix of that single column using this code:

```python
Alignment_Dummy = pd.get_dummies(df['Alignment'])
```

Then we'll show our new matrix by calling AlignmentDummy

```python
Alignment_Dummy
```



````python
Alignment_Dummy = pd.get_dummies(df['Alignment'])
````

````python
Alignment_Dummy
````

If you would like to change the column name - 
to something more descriptive, you can use the code below to change the column name Unknown.

````python
Alignment_Dummy.rename(columns = {"-":"unknown"}, inplace=True)
````

````python
Alignment_Dummy
````

If you want to dummy code more than one variable or all categorical variables at one time, you can! The syntax is a follows:

```python
# All categorical variables
df = pd.get_dummies(df)

# Multiple categorical variables 
df_Multiple = pd.get_dummies(df, columns = ['Gender','Alignment'])

# prefix Gender
df_dummies = pd.get_dummies(df, prefix='Gender', prefix_sep='_', 
             columns=['sex'])

```

Lets pull another category

````python
Gender_Dummy = pd.get_dummies(df['Gender'])
````

````python
Gender_Dummy
````

````python
Gender_Dummy.rename(columns = {"-":"unknown"}, inplace=True)
````

````python
Gender_Dummy
````
