# Part II: ANOVAs in Python
Please answer the following question.

Does the total volume of avocados sold differ between Indianapolis, Orlando, and PhoenixTucson?

```python
avocados.head()
```

```python
avocados
```

```python
# so put ['Indianapolis', 'Orlando', 'PhoenixTucson'] into a list called Region
Region = ['Indianapolis', 'Orlando', 'PhoenixTucson']
```

```python
Region
```

```python
# took your list and checked to see if it is a region in avacodoes, then put the true/false table into avocados1
avocados1 = avocados['region'].isin(Region)
```

```python
avocados1
```

```python
#copied into True and put into avocados2
avocados2 = avocados[avocados1].copy()
```

```python
#show that the boolean worked
avocados2
```

```python
#group dataset by region and get the mean total volumen
avocados2.groupby('region').mean()['Total Volume']
```

```python
# replacing the data into 0,1,2 insert into new column called RegionsNum
avocados2['RegionsNum'] = avocados2['region'].replace(['Indianapolis', 'Orlando', 'PhoenixTucson'],[0,1,2])
```

```python
#calling the dataframe to see the results
ava  = avocados2[['region','Total Volume']]
```

```python
ava.hist()
```

```python
avocados3 = avocados2[['Total Volume', 'RegionsNum']]
```

```python
avocados3
```

```python
sns.distplot(avocados3['Total Volume'])
```

```python
scipy.stats.bartlett(avocados3['RegionsNum'], avocados3['Total Volume'])
```
