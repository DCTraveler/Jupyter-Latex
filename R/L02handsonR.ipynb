```r
library(ggplot2)
```

```r
df <- read.csv("../data/Seattle_ParksnRec.csv")
```

```r
colnames(df)
```

# of trips Fall

```r
ggplot(df, aes(x=X..of.trips.Fall)) + geom_histogram(aes(y=..density..), bins = 16, colour="black", fill="white") +
        geom_density(alpha=.2, fill="#FF6666") 
```

```r
ggplot(df, aes(x=X..of.trips.Fall^(1/3))) + geom_histogram(aes(y=..density..), bins = 16,colour="black", fill="white") +
        geom_density(alpha=.2, fill="#FF6666") 
```
